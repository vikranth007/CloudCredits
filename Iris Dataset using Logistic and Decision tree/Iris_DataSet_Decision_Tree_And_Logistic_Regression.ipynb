# Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import  LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
import warnings
warnings.filterwarnings('ignore')
```

```python
iris = load_iris()
```

```python
iris.DESCR
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
X = iris.data
y = iris.target
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

# Logistic Regression

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(numeric_only=True),annot=True)
plt.show()
```

```python
cm = confusion_matrix(y_test,y_pred)
```

```python
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - Iris")
plt.show()
```

```python
print("Confusion Matrix: ",confusion_matrix(y_test,y_pred))
print('Accuracy: ',accuracy_score(y_test,y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

# Decision Tree Classifier

```python
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
y_pred_dt = dt.predict(X_test)
```

```python
cm1 = confusion_matrix(y_test,y_pred_dt)
```

```python
sns.heatmap(cm1, annot=True, fmt='d', cmap='Greens', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - Iris")
plt.show()
```

```python
print("Confusion Matrix: ",confusion_matrix(y_test,y_pred_dt))
print('Accuracy: ',accuracy_score(y_test,y_pred_dt))
print("\nClassification Report:\n", classification_report(y_test, y_pred_dt))
```
